```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Initialize the corresponding tokenizer
tokenizer = AutoTokenizer.from_pretrained(
    "meta-llama/Llama-2-7b-chat-hf", #Llama-2-Chat are optimized for dialogue use cases
)
# Load the pretrained model
model = AutoModelForCausalLM.from_pretrained(
    "meta-llama/Llama-2-7b-chat-hf",
    torch_dtype=torch.float16,
    device_map="auto", # Choose the device automatically
    trust_remote_code=False,
)
```

```python
# create prompt
prompt = "[INST] Please write a short story about a lama[/INST]"

# Inference
with torch.no_grad(): #Disable gradient calculation to reduce memory
    token_ids = tokenizer.encode(prompt, return_tensors="pt") #to return PyTorch tensors
    output_ids = model.generate(
        token_ids.to(model.device), # for cuda
        temperature=0.1, #Controls the level of randomness
        do_sample=True, #multiple possible outputs
        top_p=0.95, # probability cutoff
        top_k=30, # Top-k sampling cutoff
        max_new_tokens=256,# Maximum number of tokens to generate
    )

output = tokenizer.decode(output_ids[0][token_ids.size(1) :])# Decode the generated output
print(output)
```
